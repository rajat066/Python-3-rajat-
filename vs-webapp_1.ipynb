```python
##calculator
from tkinter import *
import tkinter.messagebox
root=Tk()
root.title("simple claculator")
root.geometry('300x450')
e=Entry(root, width=35, borderwidth=5)
e.grid(row=0, column=0, columnspan=3, padx=10, pady=10)
##e.insert(0, "enter your name")
def btn_click(number):
    current=e.get()
    e.delete(0, END)
    e.insert(0, str(current) + str(number))
def btn_clear():
    e.delete(0, END)
def btn_add():
    first_number = e.get()
    global f_num
    global math
    math="addition"
    f_num = int(first_number)
    e.delete(0, END)
def btn_equal():
    second_number=e.get()
    e.delete(0, END)
    if math == "addition":
        e.insert(0, f_num + int(second_number))
    if math == "substraction":
        e.insert(0, f_num - int(second_number))
    if math == "mul":
        e.insert(0, f_num * int(second_number))
    if math == "div":
        e.insert(0, f_num / int(second_number))
def btn_substract():
    first_number = e.get()
    global f_num
    global math
    math="substraction"
    f_num = int(first_number)
    e.delete(0, END)
def btn_product():
    first_number = e.get()
    global f_num
    global math
    math="mul"
    f_num = int(first_number)
    e.delete(0, END)
def btn_devide():
    first_number = e.get()
    global f_num
    global math
    math="div"
    f_num = int(first_number)
    e.delete(0, END)
#define buttons
btn1=Button(root, text="1", padx=40, pady=20, bg="red", fg="white", command=lambda: btn_click(1))
btn2=Button(root, text="2", padx=40, pady=20, bg="red", fg="white", command=lambda: btn_click(2))
btn3=Button(root, text="3", padx=40, pady=20, bg="red", fg="white", command=lambda: btn_click(3))
btn4=Button(root, text="4", padx=40, pady=20, bg="red", fg="white", command=lambda: btn_click(4))
btn5=Button(root, text="5", padx=40, pady=20, bg="red", fg="white", command=lambda: btn_click(5))
btn6=Button(root, text="6", padx=40, pady=20, bg="red", fg="white", command=lambda: btn_click(6))
btn7=Button(root, text="7", padx=40, pady=20, bg="red", fg="white", command=lambda: btn_click(7))
btn8=Button(root, text="8", padx=40, pady=20, bg="red", fg="white", command=lambda: btn_click(8))
btn9=Button(root, text="9", padx=40, pady=20, bg="red", fg="white", command=lambda: btn_click(9))
btn0=Button(root, text="0", padx=40, pady=20, bg="red", fg="white", command=lambda: btn_click(0))
btn_add=Button(root, text="+", padx=39, pady=20, bg="red", fg="white", command=btn_add)
btn_equal=Button(root, text="=", padx=90, pady=20, bg="red", fg="white", command=btn_equal)
btn_clear=Button(root, text="c", padx=90, pady=20, bg="red", fg="white", command=btn_clear)
btn_substract=Button(root, text="-", padx=40, pady=20, bg="red", fg="white", command=btn_substract)
btn_product=Button(root, text="*", padx=40, pady=20, bg="red", fg="white", command=btn_product)
btn_devide=Button(root, text="/", padx=40, pady=20, bg="red", fg="white", command=btn_devide)
#put the buttons on the screen
btn1.grid(row=3, column=0)
btn2.grid(row=3, column=1)
btn3.grid(row=3, column=2)
btn4.grid(row=2, column=0)
btn5.grid(row=2, column=1)
btn6.grid(row=2, column=2)
btn7.grid(row=1, column=0)
btn8.grid(row=1, column=1)
btn9.grid(row=1, column=2)
btn0.grid(row=4, column=0)
btn_clear.grid(row=4, column=1, columnspan=2)
btn_add.grid(row=5, column=0)
btn_equal.grid(row=5, column=1, columnspan=2)
btn_substract.grid(row=6, column=0)
btn_product.grid(row=6, column=1)
btn_devide.grid(row=6, column=2)
root.mainloop()
```

```python
##image viewer with status
from tkinter import *
from PIL import ImageTk, Image
root=Tk()
root.title('welcome to the window')
root.iconbitmap('C:/Users/rajat/Desktop/python/lab/car.png')
my_img1=ImageTk.PhotoImage(Image.open("C:/Users/rajat/Desktop/python/lab/logo.png"))
my_img2=ImageTk.PhotoImage(Image.open("C:/Users/rajat/Desktop/python/lab/maxican.jpg"))
my_img3=ImageTk.PhotoImage(Image.open("C:/Users/rajat/Desktop/python/lab/ram.jpg"))
my_img4=ImageTk.PhotoImage(Image.open("C:/Users/rajat/Desktop/python/lab/sulvin.jpg"))

image_list=[my_img1, my_img2, my_img3, my_img4]
status=Label(root, text="Image 1 of" + str(len(image_list)), bd=1, relief=SUNKEN, anchor=W)

lbl1=Label(image=my_img1)
lbl1.grid(row=0, column=0, columnspan=3)

def forward(image_no):
    global lbl1
    global btn_forward
    global btn_back
    lbl1.grid_forget()
    lbl1=Label(image=image_list[image_no-1])
    btn_forward=Button(root, text='>>', command=lambda: forward(image_no+1))
    btn_back=Button(root, text='<<', command=lambda: back(image_no-1))
    
    if image_no == 4:
        btn_forward=Button(root, text=">>", state=DISABLED)
    
    lbl1.grid(row=0, column=0, columnspan=3)
    btn_forward.grid(row=1, column=2)
    btn_back.grid(row=1, column=0)
    
    status=Label(root, text="Image " + str(image_no) + "of" + str(len(image_list)), bd=1, relief=SUNKEN, anchor=W)
    status.grid(row=2, column=2, columnspan=3, sticky=W+E)

def back(image_no):
    global lbl1
    global btn_forward
    global btn_back
    lbl1.grid_forget()
    lbl1=Label(image=image_list[image_no-1])
    btn_forward=Button(root, text='>>', command=lambda: forward(image_no+1))
    btn_back=Button(root, text='<<', command=lambda: back(image_no-1))
    
    if image_no == 1:
        btn_back = Button(root, text='<<', state=DISABLED)
    
    lbl1.grid(row=0, column=0, columnspan=3)
    btn_forward.grid(row=1, column=2)
    btn_back.grid(row=1, column=0)
    
    status=Label(root, text="Image " + str(image_no) + "of" + str(len(image_list)), bd=1, relief=SUNKEN, anchor=W)
    status.grid(row=2, column=2, columnspan=3, sticky=W+E)
    
btn_back=Button(root, text='<<', command=back, state=DISABLED)
btn_quit=Button(root, text='Exit', command=root.destroy)
btn_forward=Button(root, text='>>', command=lambda: forward(2))

btn_back.grid(row=1, column=0)
btn_quit.grid(row=1, column=1)
btn_forward.grid(row=1, column=2, pady=10)
status.grid(row=2, column=2, columnspan=3, sticky=W+E)

root.mainloop()
```
